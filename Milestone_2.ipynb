### Correlation Tests

Now that we have a summary of the data, we will run some correlation tests. We want to explore how well various statistics (runs scored, runs allowed, OBP, SLG, BA) can predict Wins / Playoff Appearances for MLB teams.  Our initial hypothesis is that OBP will be a better predictor of Wins / Playoff Appearances than either SLG or BA.

---Remember correlation is not causation---

Correlation Scale:

0 - no association

0 - +/-.25 = negligible

+/-.25 - +/-.5 = weak

+/-.5 - +/-.75 - moderate

+/-.75 - +/-1 = strong

```python
#Creating a dataframe of Runs Saved and OBP
OBP_corr_df = ['RS', 'OBP']
OBP_corr = baseball_df[OBP_corr_df]
```

```python
#Creating a master table of all correlations 
correlations_master = baseball_df.drop(['Year'],axis=1).corr(method='pearson')
correlations_master
```

```python
#Creating a clustermap to show correlations between the previously mentioned categories and dropping unnecessary categories.
#Correlation values are shown and the colors correspond to the provided key.
def draw_cluster(df, fmt=None):
    correlations_drop_nan = baseball_df.drop(['Year', 'RankSeason', 'RankPlayoffs','OOBP','OSLG','G'],axis=1).corr(method='pearson')
    master_cluster = sns.clustermap(correlations_drop_nan, cmap='Spectral', linewidth=.5, metric='correlation', annot=(fmt!=None),fmt=fmt)
    plt.setp(master_cluster.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
    plt.show()

draw_cluster(baseball_df,'.3f')
```

```python
#Creating a heatmap similar to the above with more categories shown.
#strong correlations in dark blue, strong negative correlations in white.
sns.heatmap(baseball_df.drop(['Year'],axis=1).corr(),cmap='Blues',annot=False) 
```

<div style="background-color:LightGrey;">
    
### Correlation Results

Breaking down the correlation results for each category.

RS: strong correlations(in order of stronger to weaker):SLG,OBP,BA; surprisingly weak postitive correlation with W, Playoffs, & Ranks

RA: appropriately weak correlations with W, OBP, SLG, BA, Playoffs, and Ranks

W: weak correlations with OBP, SLG, and BA; slightly higher for Playoffs; strong negative correlation with RankSeason 

OBP: strong correlations with RS, SLG, BA; weak for Playoffs, and Rank

SLG: strong correlations with RS, OBP, and BA; weak for Playoffs and Rank

BA: strong correlations with RS, OBP, and SLG; weak for Playoffs and Rank

STRONGEST CORRELATIONS (strongest to weakest +-.5 or higher):

RS: SLG, OBP, BA, W

W: Playoffs, RS, (-) RA, RankSeason

Playoffs: W

RankSeason: (-) W.

------------------------------------------
As expected, all of the offensive production categories (RS, SLG, BA, OBP) show strong correlations with each other. 
We were hoping to find strong correlations between specific hitting categories and Wins / Playoffs. 

Notably, the correlation between Wins and RankSeason is negative.  This actually makes perfect sense, as the best rank (1) is the lowest possible value for that variable and, in any given season, goes to the team with the most wins.  While wins are an individual statistic given to pitchers, the team statistic wins is equal by identity to the sum of the wins recorded by all of that team's pitchers.

```python
#pair plots
sns.pairplot(baseball_df, hue = "Playoffs", vars=['RS','OBP','SLG','BA'])
#teams that made the playoffs were better across all offensive categories, but no single offensive category stood out as
#being better than the others.
```

```python
#Calculating R^2.  Measures the proportion of variance in Y explained by X - e.g. how much does RS explain the variance
#in W.  The result tells us that 26.2% of the variance in W can be explained by RS.
baseball_df['W'].corr(baseball_df['RS'], method='pearson')**2
```

```python
#Hypothesis: teams that score more runs win more games
baseball_df['W'].corr(baseball_df['RS'], method='pearson')**2
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(6,6))
axes.scatter(baseball_df["RS"],baseball_df["W"], color = "dodgerblue")
axes.grid(alpha=0.25)
axes.set_xlabel("Runs Scored", fontsize=16)
axes.set_ylabel("Wins", fontsize=16)
axes.set_title("Wins vs. Runs Scored", fontsize=16)
#There appears to be a positive relationship between RS and W as evidenced by the below scatterplot and the moderately 
#strong correlation between the two variables.  The outcome makes sense and it further makes sense that the explanatory
#power is not higher than 26% for at least two reasons.  First, a team's run prevention is likely as important to wins
#as its runs scored.  Second, the relationship between runs scored and wins varies strongly depending upon run environment.
#For example, some seasons (such as those taking place during baseball's so-called "PED Era" in the 1990's and early 200's)
#saw teams regularly score far more runs than other seasons (such as those in the late 1960's when various rule changes
#benefited pitchers).  Further, some teams (such as Colorado) play in ballparks that are more conducive to run-scoring than
#others (such as the LA Dodgers).  These run environment factors explain some of the variance in wins that runs scored and
#runs allowed to not fully explain.
```

```python
#Hypothesis: teams that have higher OBP win more games
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(6,6))
axes.scatter(baseball_df["OBP"],baseball_df["W"], color = "mediumspringgreen")
axes.grid(alpha=.25)
axes.set_xlabel("OBP", fontsize=16)
axes.set_ylabel("Wins", fontsize=16)
axes.set_title("Wins vs. OBP", fontsize=16)

#The scatter plot below shows a clear positive correlation between team wins and OBP.  As shown in the above correlations
#matrix, the correlation between OBP and wins is 0.481836.  By comparison, the correlations between wins and both BA and SLG
#are both approximately 0.40.
```
